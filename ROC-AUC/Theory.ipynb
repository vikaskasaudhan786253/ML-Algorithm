### ROC (Receiver Operating Characteristic)
1. #### Why do we use ROC in ML ?
> used to evaluate `classification models`,especially `binary classifiers`

2. #### Why ROC is important ?

   * Threshold-independent evaluation
   * works well with imbalanced data.
   * Compares multiple models visually.
   * focuses on errors that matters.

3. #### What is ROC ?
   > It is a graph between `TPR`(y-axis) and `FPR`(x-axis) for different thresholds.
   
   > TPR = TP/(TP + FN)

   > FPR = FP/(FP + TN)

4. #### AUC (Area Under Curve)

   > AUC = 1.0 -> Perfet Classifier.

   > AUC = 0.5 -> Random Guessing.

   > Higher AUC = Better Model.
