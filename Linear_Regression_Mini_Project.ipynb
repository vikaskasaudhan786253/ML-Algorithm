```python
import pandas as pd
```

```python
df1 = pd.read_csv('jee_2023.csv')
df2 = pd.read_csv('jee_2024.csv')
df3 = pd.read_csv('jee_2025.csv')
df = pd.concat([df1,df2,df3],ignore_index=True)
```

### `Q` Chapter Weightage Trend Analyzer

`columns`:
* year
* chapter name
* number of question

```python
temp = df.copy()
```

```python
temp.isnull().sum()
```

```python
temp = temp[['date','subject_name','chapter_name','topic_name','shift_number']]
```

```python
temp
```

```python
temp.info()
```

```python
new_df = temp.dropna()
```

```python
new_df.info()
```

```python
new_df.duplicated().sum()
```

```python
temp_df = new_df.drop_duplicates()
```

```python
temp_df
```

```python
temp.info()
```

```python
temp_df['subject_name'] = temp_df['subject_name'].apply(lambda x:x.title())
temp_df['chapter_name'] = temp_df['chapter_name'].apply(lambda x:x.title())
temp_df['topic_name'] = temp_df['topic_name'].apply(lambda x:x.title())
temp_df['shift_number'] = temp_df['shift_number'].apply(lambda x:x.title())
```

```python
temp_df['date'].apply(lambda x:x)
```

```python
temp_df['date'].unique()
```

```python
temp_df['date'] = pd.to_datetime(temp_df['date'],format='mixed')
```

```python
temp_df.info()
```

```python
temp_df['year'] = temp_df['date'].dt.year
```

```python
new_temp_df = temp_df
```

```python
p1 = new_temp_df.groupby(['year','chapter_name'])['shift_number'].count().sort_values(ascending=False).reset_index()
```

```python
p2 = new_temp_df.groupby(['year','chapter_name','topic_name'])['shift_number'].count().sort_values(ascending=False).reset_index()
```

```python
p1 = p1.rename(columns={'shift_number':'no_of_questions'})
```

```python
p2 = p2.rename(columns={'shift_number':'no_of_questions'})
```

```python
p1
```

```python
p2
```

```python
# no_of_questions = B0 + B*(year)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = p1.iloc[:,0:1]
y = p1.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
x_train , x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import numpy as np
```

```python
print('MAE',mean_absolute_error(y_test,y_pred))
print('MSE',mean_squared_error(y_test,y_pred))
print('MSE',np.sqrt(mean_squared_error(y_test,y_pred)))
print('R2 Score',r2_score(y_test,y_pred))
```
