```python
from mlxtend.evaluate import bias_variance_decomp
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from mlxtend.data import boston_housing_data
from sklearn.model_selection import train_test_split
```

```python
x, y = boston_housing_data()
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=123,
                                                 shuffle=True)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lr = LinearRegression()

avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
                                  lr, x_train, y_train, x_test,y_test,
                                  loss='mse', random_seed=123)

print('Average expected loss: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance %.3f' % avg_var)
```

```python
dt = DecisionTreeRegressor(random_state=123)

avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
                                       dt,x_train,y_train,x_test,y_test,
                                       loss='mse',random_seed=123)

print('Average expected loss %.3f' % avg_expected_loss)
print('Average bias %.3f' % avg_bias)
print('Average variance %.3f' % avg_var)
```

# 1. What is Bias?

     *  Bias is the error due to wrong assumptions in the learning algorithm.
     
     * or bias is inability of a machine learning model to fit the training data.
     
     * or It measures how far the model’s average prediction is from the true function.

### Characteristics of High Bias

* Model is too simple

* Cannot capture complex patterns

* Strong assumptions about data

* Causes underfitting


### Examples of High Bias Models

* Linear regression on non-linear data

* Shallow decision tree

* KNN with very large K

# 2. What is Variance?

     * Variance is the error due to too much sensitivity to training data.

     * or  It measures how much predictions change if training data changes.

### Characteristics of High Variance

* Model is too complex

* Learns noise in data

* Performs very well on training data

* Performs poorly on test data

* Causes overfitting

### Examples of High Variance Models

* High-degree polynomial regression

* Deep decision trees

* KNN with very small K

* Neural networks without regularization


### Total Error = (Bias)^2 + Variance + Irreducible Error
                >  underfitting  +  overfitting  +  noise

* `irreducible error ` can not be eliminated

### 9. How to Reduce Bias?

Use when underfitting

* Increase model complexity

* Add more features

* Reduce regularization

* Use non-linear models

* Train longer (for NN)

### 10. How to Reduce Variance?

Use when overfitting

* Collect more training data

* Feature selection

* Increase regularization (L1/L2)

* Early stopping

* Pruning (trees)

* Dropout (neural networks)

* Bagging / Random Forest
