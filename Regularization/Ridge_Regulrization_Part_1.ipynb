```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.linear_model import Ridge

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = load_diabetes()
```

* `DESCR` --- detailed information about the dataset means `description`
* `random_state` is a seed value for the random number generator used while shuffling and splitting the data.
* It controls `randomness` so that we get the same train–test split every time we run the code.
* If the `random_state` value `changes`, the `randomness` changes.

```python
print(data.DESCR)
```

```python
x = data.data
y = data.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
L = LinearRegression()
L.fit(x_train,y_train)
```

```python
print(L.coef_)
print(L.intercept_)
```

```python
y_pred = L.predict(x_test)
```

```python
print("R2 score: ",r2_score(y_test,y_pred))
print("RMSE: ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
R = Ridge(alpha=100000)
R.fit(x_train,y_train)
```

```python
print(R.coef_)
print(R.intercept_)
```

```python
y_pred1 = R.predict(x_test)
```

```python
print("R2 score: ",r2_score(y_test,y_pred1))
print("RMSE: ",np.sqrt(mean_squared_error(y_test,y_pred1)))
```

#### This code fits polynomial ridge regression models with different alpha values and visualizes how regularization controls overfitting.

```python
def get_preds_ridge(x1,x2,alpha):
    model = Pipeline([
        ('poly_feats',PolynomialFeatures(degree=16)),
        ('ridge',Ridge(alpha=alpha))
    ])
    model.fit(x1,x2)
    return model.predict(x1)

alphas = [0, 20, 200]
cs = ['r', 'g', 'b']

plt.figure(figsize=(10,6))
plt.plot(x1,x2, 'b+', label='Datapoints')

for alpha, c in zip(alphas,cs):
    preds = get_preds_ridge(x1,x2,alpha)
    #plot
    plt.plot(sorted(x1[:,0]), preds[np.argsort(x1[:,0])], c,
            label='Alpha:{}'.format(alpha))

plt.legend()
plt.show()
```
